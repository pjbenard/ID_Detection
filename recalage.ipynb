```python
import numpy as np

from PIL import Image

import matplotlib.pyplot as plt
```

```python
img = np.array(Image.open('CNI_example.jpg'))
mask = np.array(Image.open('mask_example.jpg'))
img.shape, mask.shape
```

```python
mask[mask > 127] = 255
mask[mask < 127] = 0
```

```python
def mask_array(arr, mask, fill_value=0):
    masked_arr = np.where(mask, arr, fill_value)
    return masked_arr
```

```python
img_masked = mask_array(img, mask)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 6), sharey=True, )
axs[0].imshow(img)
axs[1].imshow(mask)
axs[2].imshow(img_masked)
```
